```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('All_IPL_Match_Scores_easy.xlsx')
df.head()
```

```python
a = df[(df['Winner'] == 'Mumbai') & (df['Winning_Team'] == 'FirstBatting')].groupby('Year').Match_Number.count()
a
```

```python
plt.xlabel( 'Mumbai won number')
plt.ylabel('Year')
plt.plot(a,color = "green",marker='*')
```

```python
b = df[(df['Winner'] == 'Bangalore') & (df['Winning_Team'] == 'FirstBatting')].groupby('Year').Match_Number.count()
b
```

```python
plt.xlabel( 'Banglore won number')
plt.ylabel('Year')
plt.plot(b,color = "blue",marker='*')
plt.plot(a,color = "green",marker='*')
```

```python
c = df[(df['Winner'] == 'Pune') & (df['Winning_Team'] == 'FirstBatting')].groupby('Year').Match_Number.count()
c
```

```python
plt.xlabel( 'won number')
plt.ylabel('Year')
plt.plot(a,color = "green",linestyle='dashdot')
plt.plot(b,color = "blue",marker='*')
plt.plot(c,color = "red",linestyle='-')
```

```python
d = df[(df['Winner'] == 'Chennai') & (df['Winning_Team'] == 'FirstBatting')].groupby('Year').Match_Number.count()
d
```

```python
plt.xlabel( 'won number')
plt.ylabel('Year')
plt.hist(d,color = "brown",edgecolor='black')
```
