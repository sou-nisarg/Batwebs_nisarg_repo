```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd. read_csv("USA_Housing.csv")
df.head()
```

```python
df.columns
```

```python
x=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Avg. Area Number of Bedrooms', 'Area Population']]
y=df['Price']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
x
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.9,random_state=1)
```

```python
xtrain
```

```python
ytrain
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
col= ['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Avg. Area Number of Bedrooms', 'Area Population']
coeff_df = pd.DataFrame(model.coef_,col,columns=['M'])
coeff_df['c']=model.intercept_
coeff_df
```

```python
ypred=model.predict(xtest)
ypred
```

```python
ytest.values
```

```python
from sklearn.metrics import r2_score
r2=r2_score(ytest,ypred)
r2
```
